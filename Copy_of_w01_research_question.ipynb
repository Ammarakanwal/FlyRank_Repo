# My Lane

I selected **Refresh / Content Opportunity Scoring** as my project lane.

The objective of this project is to identify webpages that should be reviewed first for content refresh or optimization. Since organizations often have thousands of webpages but limited time and resources, prioritizing the right pages can help improve search visibility more efficiently.

I selected this lane because it combines business decision-making with data analysis and machine learning to support evidence-based SEO recommendations.

# Research Question

## Search Question

Can historical SEO and engagement metrics be used to identify webpages that should be prioritized for content refresh?

## Decision

Which webpages should an SEO team review first?

## Unit of Analysis

One webpage (one content page).

## Output

A ranked list of webpages according to their refresh priority.

## Action

The SEO team can refresh, improve, expand, monitor or optimize the highest-ranked pages first.

## Cost of a Wrong Recommendation

If an important page is not selected, valuable traffic and business opportunities may be lost.

If an unimportant page is selected, time and resources will be wasted.

## Why Data or Machine Learning?

The dataset contains many measurable SEO indicators including impressions, clicks, CTR, average position, engagement rate and content age. Instead of manually reviewing every webpage, machine learning can recognize useful patterns and help prioritize pages more efficiently.

# Quick Look at the Data

# Careful Words

This project analyzes historical SEO performance data to identify patterns associated with webpages that may require content review.

The results should not be interpreted as proof that refreshing a webpage will automatically increase traffic. Search performance can also be influenced by Google's algorithms, competition, seasonality and user behavior.

Therefore, the recommendations produced by this project are intended to support human decision-making rather than replace it.
